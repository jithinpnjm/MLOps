```python
import string
def password_ckecker(password):
    length=len(password)
    lower_case= any(c.islower() for c in password)
    upper_case= any(c.isupper() for c in password)
    special_char=any(c in string.punctuation for c in password)
    print([c for c in password if c in string.punctuation])
    return(length and lower_case and upper_case and special_char)
passwd=input("Eter your password to check")
print(f"Is password Valid: {password_ckecker(passwd)}")
```

```python
def factorial(num):
    if num==0:
        return 1
    return (num * factorial(num-1))
print(factorial(5))
```

```python
def palindrome(text):
    text=text.replace(" ","")
    if text==text[::-1]:
        return True
    else:
        return False
palindrome("level aba")
```

```python
def palindrome_new(text):
    text=text.replace(" ","")
    n=len(text)
    for i in range(n):
        if text[i]!=text[n-1-i]:
            return False
            break
    return True   
 
palindrome_new("level level")           
```

```python
## Bank account using Class and Objects
class BankAccount:
  def __init__(self,name,balance=0):
        self.name=name 
        self.balance=balance
  def BlanceCheck(self):
      print(f"The current balance is {self.balance}")
  def Deposit(self,amount):
      self.balance+=amount
      print(f"Deposit of {amount} is success. And current balance is {self.balance}")
  def Withdraw(self,amount):
      self.balance-=amount
      print("The balance is insufficent") if amount>=self.balance else print("processing withdrawal of amount {amount}")
      print (f"Amount \
      withdrawal successful and new balance is {self.balance}")

user1=BankAccount("Jithin")
user1.BlanceCheck()
user1.Deposit(100)
user1.Withdraw(99)
user1.BlanceCheck()
```

```python
import mlflow
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

# ── Connect to the tracking server ──────────────────────────
mlflow.set_tracking_uri("sqlite:///MLFLOW/mlflow/mlflow.db")
print(f"Connected to tracking URI: {mlflow.get_tracking_uri()}")

# ── Set the experiment (creates it if it doesn't exist) ────
mlflow.set_experiment("Test Experiment")

# ── Prepare data ─────────────────────────────────────────────
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(
    iris.data, iris.target, test_size=0.2, random_state=42
)

# ── Start a run and log everything ───────────────────────────
with mlflow.start_run(run_name="My First Run") as run:

    # 1. Log parameters (your model's config inputs)
    mlflow.log_param("param", 5)
    mlflow.log_params({"param1": 10, "param2": 26})

    # 2. Train a simple model
    model = LogisticRegression(max_iter=200)
    model.fit(X_train, y_train)
    accuracy = model.score(X_test, y_test)

    # 3. Log metrics (your model's output results)
    mlflow.log_metric("metric1", 0.85)
    mlflow.log_metrics({"metric2": 0.98, "metric3": 89.5})
    mlflow.log_metric("accuracy", accuracy)

    # 4. Add a tag (free-form metadata)
    mlflow.set_tag("Owner", "jithin")

    # 5. Save the trained model as an artifact
    #    Note: log_model(model_object, artifact_folder_name)
    mlflow.sklearn.log_model(model, "model")

    # 6. mlflow.log_artifact() logs an EXISTING file from disk.
    #    We skip confusion_matrix.png and model_report.txt here
    #    because those files don't exist yet.
    #    You will generate and log them in the next notebooks.

    print(f"Run ID    : {run.info.run_id}")
    print(f"Accuracy  : {accuracy:.4f}")
    print(f"Experiment: Test Experiment")
    print("✅ Run logged successfully! Check the MLflow UI.")
```
